## Objective
To understand how encryption hides a message using a key.

## You will need
- A shared book (everyone must use the same one)  
- Paper and pen  

***

## Step 1: Agree the key
The **key** is:  
- the **book title**  
- the **page number**  
- the **line number**  
- the **word number**

> Example key: *Page 12, Line 3, Word 5*

> **Tip:** Without the key, the message cannot be read.

***

## Step 2: Encrypt a message
1. Write a short message (e.g. **MEET AT LUNCH**).

## Step 3: Decrypt the message
